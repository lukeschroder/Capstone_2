```python
def gentype(row):
    if row['DATAYEAR'] >= 2014:
        if row['TECHNOLOGY'] == 'Coal Integrated Gasification Combined Cycle' or row['TECHNOLOGY'] == 'Conventional Steam Coal':
            return 'coal'
        elif row['TECHNOLOGY'] == "Natural Gas Fired Combined Cycle" or row['TECHNOLOGY'] == 'Natural Gas Fired Combustion Turbine' or row['TECHNOLOGY'] == 'Natural Gas Internal Combustion Engine' \
        or row['TECHNOLOGY'] == 'Natural Gas Steam Turbine' or row['TECHNOLOGY'] == 'Natural Gas with Compressed Air Storage' or row['TECHNOLOGY'] == 'Other Natural Gas' or row['TECHNOLOGY'] == 'Other Gases'or row['TECHNOLOGY'] == 'Landfill Gas':
            return 'natural_gas'
        elif row['TECHNOLOGY'] == 'Petroleum Coke' or row['TECHNOLOGY'] == 'Petroleum Liquids':
            return 'oil'
        elif row['TECHNOLOGY'] == 'Conventional Hydroelectric' or row['TECHNOLOGY'] == 'Hydroelectric Pumped Storage':
            return 'hydroelectric'
        elif row['TECHNOLOGY'] == 'Geothermal':
            return 'geothermal'
        elif row['TECHNOLOGY'] == 'Onshore Wind Turbine' or row['TECHNOLOGY'] == 'Offshore Wind Turbine':
            return 'wind'
        elif row['TECHNOLOGY'] == 'Solar Photovoltaic' or row['TECHNOLOGY'] == 'Solar Thermal with Energy Storage' or row['TECHNOLOGY'] == 'Solar Thermal without Energy Storage':
            return 'solar'
        elif row['TECHNOLOGY'] == 'Nuclear':
            return 'nuclear'
        else:
            return 'other/unreported'
    else:
        if row['ENERGYSOURCE1'] == 'ANT' or row['ENERGYSOURCE1'] == 'BIT' or row['ENERGYSOURCE1'] == 'COL' or row['ENERGYSOURCE1'] == 'COM' or row['ENERGYSOURCE1'] == 'CWM' or row['ENERGYSOURCE1'] == 'SUB'\
        or row['ENERGYSOURCE1'] == 'LIG' or row['ENERGYSOURCE1'] == 'SGC' or row['ENERGYSOURCE1'] == 'WC' or row['ENERGYSOURCE1'] == 'RC':
            return 'coal'
        elif row['ENERGYSOURCE1'] == 'BFG' or row['ENERGYSOURCE1'] == 'NG' or row['ENERGYSOURCE1'] == 'OG' or row['ENERGYSOURCE1'] == 'GAS' or row['ENERGYSOURCE1'] == 'MTH' or row['ENERGYSOURCE1'] == 'LNG'\
        or row['ENERGYSOURCE1'] == 'LPG' or row['ENERGYSOURCE1'] == 'RG' or row['ENERGYSOURCE1'] == 'MTH' or row['ENERGYSOURCE1'] == 'SNG':
            return 'natural_gas'
        elif row['ENERGYSOURCE1'] == 'DFO' or row['ENERGYSOURCE1'] == 'JF' or row['ENERGYSOURCE1'] == 'KER' or row['ENERGYSOURCE1'] == 'PC' or row['ENERGYSOURCE1'] == 'PG' or row['ENERGYSOURCE1'] == 'RFO'\
        or row['ENERGYSOURCE1'] == 'SGP' or row['ENERGYSOURCE1'] == 'WO' or row['ENERGYSOURCE1'] == 'CRU' or row['ENERGYSOURCE1'] == 'FO1' or row['ENERGYSOURCE1'] == 'FO2' or row['ENERGYSOURCE1'] == 'FO3'\
        or row['ENERGYSOURCE1'] == 'FO4' or row['ENERGYSOURCE1'] == 'FO5' or row['ENERGYSOURCE1'] == 'FO6' or row['ENERGYSOURCE1'] == 'JF' or row['ENERGYSOURCE1'] == 'KER' or row['ENERGYSOURCE1'] == 'PET'\
        or row['ENERGYSOURCE1'] == 'TOP':
            return 'oil'
        elif row['ENERGYSOURCE1'] == 'WAT':
            return 'hydroelectric'
        elif row['ENERGYSOURCE1'] == 'GST' or row['ENERGYSOURCE1'] == 'GEO':
            return 'geothermal'
        elif row['ENERGYSOURCE1'] == 'WND':
            return 'wind'
        elif row['ENERGYSOURCE1'] == 'SUN':
            return 'solar'
        elif row['ENERGYSOURCE1'] == 'NUC' or row['ENERGYSOURCE1'] == 'UR':
            return 'nuclear'
        else:
            return 'other/unreported'

sixtys_agg['GEN_TYPE'] = sixtys_agg.apply (lambda row: gentype(row), axis=1)
```
